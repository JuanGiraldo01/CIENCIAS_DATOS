###creacion de dataframes por medios de diccionarios

```python
elementos={
    "numero atomico":[1,6,47,88],
    "masa atomica":[1,12,107,226],
    "familia":["no metal","no metal","metal","metal"]
}
elementos
```

```python
tabla_periodica=pd.DataFrame(elementos)
tabla_periodica
```

```python
tabla_periodica=pd.DataFrame(elementos,index=["H","C","Ag","Ra"],columns=["familia","numero atomico","masa atomica"])
tabla_periodica
```

###Creando un dataframe a partir de un array Numpy

```python
datos=np.array([[2,4,6,8],[4,6,7,4],[5,2,3,1]])
datos
```

```python
unidades=pd.DataFrame(datos)
unidades
```

```python
unidades.index=["la ureña","la predregoza","maria bonita"]
unidades.columns=[2021,2022,2023,2024]
unidades
```

###Creacion de dataframes con diccionarios

```python
medida_2020={"Ag":2,"Au":8,"Cu":3,"pt":6}
medida_2021={"Ag":4,"Au":1,"Cu":2,"pt":2}
medida_2022={"Ag":5,"Au":2,"Cu":7,"pt":9}

unidades=pd.DataFrame([medida_2020,medida_2021,medida_2022],index=list(range(2020,2023)))
unidades
```

```python
medida_2020={"Ag":2,"Au":8,"Cu":3,"pt":6}
medida_2021={"Ag":4,"Au":1,"Cu":2,"pt":2}
medida_2022={"Ag":5,"pb":2,"Cu":7,"pt":9}

unidades=pd.DataFrame([medida_2020,medida_2021,medida_2022],index=list(range(2020,2023)))
unidades
```
